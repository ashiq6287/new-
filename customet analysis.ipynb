```python
df.drop(["Unnamed: 0.1","Unnamed: 0","Education","Kidhome","Teenhome","Dt_Customer","Recency","MntCoke","MntFruits"],axis=1,inplace=True)
```

```python
df.drop(["MntMeatProducts","MntFishProducts","MntSweetProducts","MntGoldProds","NumWebPurchases"],axis=1,inplace=True)
```

```python
df.drop(["NumCatalogPurchases","NumStorePurchases","NumWebVisitsMonth","AcceptedCmp3","AcceptedCmp4","AcceptedCmp5","AcceptedCmp1","AcceptedCmp2"],axis=1,inplace=True)
```

```python
plt.figure(1, figsize=(15,6))
n=0
for x in ['ID' , 'Year_Birth' , 'Marital_Status' , 'Income' , 'NumDealsPurchases']:
    n+=1
    plt.subplot(1 , 3 , n)
    plt.subplots_adjust(hspace =0.5 , wspace = 0.5)
    sns.distplot(df[x] , bins = 20)
    plt.title('Displot of {}' .format(x))
plt.show()  
```

```python
plt.figure(figsize=(15,5))
sns.countplot(y='Marital_Status' ,data=df)
plt.show()
```

```python
plt.figure(1,figsize=(15,7))
n=0
for cols in ['ID' , 'Year_Birth' , 'Marital_Status' , 'Income' , 'NumDealsPurchases']:
    n+=1
    plt.subplot(1 , 3, n)
    sns.set(style="whitegrid")
    plt.subplots_adjust(hspace = 0.5 , wspace = 0.5)
    sns.violinplot(x = cols , y = 'Marital_Status' , data=dp )
    plt.ylabel('Marital_Status' if n== 1 else ' ')
    plt.title('Violin Plot')
plt.show()    
```
